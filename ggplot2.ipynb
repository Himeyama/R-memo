```r
library("ggplot2")
```

```r
y <- list("Petal.Length", "Petal.Width")
x <- list("Sepal.Length", "Sepal.Width")

for(ey in y){
    for(ex in x){
        p <- ggplot(data<-iris, mapping<-aes_string(x = ex, y = ey, color="Species")) + 
#             geom_line() +
            geom_point() +
            labs(color="凡例", title="タイトル") +
            theme_linedraw() +
            theme(plot.title=element_text(hjust=0.5, size=16)) +
#             xlim(0, 10) + ylim(0, 10) +
            xlab("横") + ylab("縦")
            
        print(p)
    }
}
```
